```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler , OneHotEncoder
import numpy as np
import pickle
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
model= load_model('model.h5')
```

```python
with open('one_hot_encoder_geo.pkl','rb') as file:
    one_hot_encoded_geo = pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender = pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler = pickle.load(file)
```

```python
input_data = pd.DataFrame({
    'CreditScore': [619],
    'Geography': ['France'],
    'Gender': ['Female'],
    'Age': [42],
    'Tenure': [2],
    'Balance': [0.00],
    'NumOfProducts': [1],
    'HasCrCard': [1],
    'IsActiveMember': [1],
    'EstimatedSalary': [101348.88]
})
```

```python
# Encode gender (lowercase to match training)
input_data['gender'] = label_encoder_gender.transform(input_data['Gender'])
```

```python
print(type(one_hot_encoded_geo))
```

```python
# One-hot encode Geography
geo_encoded = one_hot_encoded_geo.transform(input_data[['Geography']]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=one_hot_encoded_geo.get_feature_names_out())
```

```python
# Build final input_df
input_df = pd.concat([input_data.drop(['Geography', 'Gender'], axis=1), geo_encoded_df], axis=1)
```

```python
input_scaled = scaler.transform(input_df)
```

```python
print(input_df.columns.tolist())
```

```python
#prediction
prediction = model.predict(input_scaled)
prediction
```
